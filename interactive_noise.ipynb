# Interactive Noise Demonstration

This notebook contains a slider widget to visually demonstrate adding noise (between 0.01% and 10%) to MNIST images.

```python
import torch
import matplotlib.pyplot as plt
from torchvision import datasets, transforms
import ipywidgets as widgets
from IPython.display import display
import os

# Set up the paths to load the local data (from config.py's directory logic)
DATA_DIR = './data'

# Download or load data
transform_clean = transforms.ToTensor()
testset = datasets.MNIST(root=DATA_DIR, train=False, download=True, transform=transform_clean)

# Get a single random image for demonstration
image, _ = testset[0]

def add_gaussian_noise(tensor, noise_factor):
    noise = torch.randn_like(tensor) * noise_factor
    noisy_image = tensor + noise
    return torch.clamp(noisy_image, 0., 1.)

def visualize_noise(noise_percentage):
    # Percentage to factor (e.g. 5% -> 0.05)
    noise_factor = noise_percentage / 100.0
    noisy_img = add_gaussian_noise(image, noise_factor)

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4))
    
    # Original Image
    ax1.imshow(image.squeeze(), cmap='gray')
    ax1.set_title('Original Clean Image')
    ax1.axis('off')
    
    # Noisy Image
    ax2.imshow(noisy_img.squeeze(), cmap='gray')
    ax2.set_title(f'Noisy Image ({noise_percentage:.2f}% Noise)')
    ax2.axis('off')
    
    plt.show()

# Create the interactive slider from 0.01% to 10%
noise_slider = widgets.FloatSlider(
    value=5.0,
    min=0.01,
    max=10.0,
    step=0.01,
    description='Noise %:',
    continuous_update=False
)

# Render
_ = widgets.interact(visualize_noise, noise_percentage=noise_slider)
```
